# poisson distribution : 

$$ Po(x|\alpha) = \frac{\alpha^x e^{-\alpha}}{x!}$$

# normal distribution : 

$$ N(x| \mu, \sigma) = \frac{1}{\sqrt{2\pi} \sigma} exp(-\frac{(x-\mu)^{2}}{2\sigma^{2}}) $$

$$ \text{we can prove that for a high enough mean ( > 20 ) the possion distribution can be approximated with a normal distribution, we can do this using two theorems from statistics :} $$

1. $\text{the central limit theorem}$

2. $\text{the limit equivalence of the moment generating function}$ 


# reasoning or proof :


to prove that poisson distribution can be approximated by a normal distribution we have to use the fact that the <b>moment generating function of the original distribution tends to moment generating function of the normal distribution</b>

$$\begin{align}
\lim_{\alpha\to\infty}M_{x}&=
\lim_{\alpha\to\infty}\Bbb{E}\left(\exp{(tx)}\right)\\
&=\lim_{\alpha\to\infty}\Bbb{E}\left(\exp{\left(t\frac{z-{\alpha}}{\sqrt{\alpha}}\right)}\right)\\
&=\lim_{\alpha\to\infty}\exp(-t\sqrt\alpha)\Bbb{E}\left(\exp{\left(\frac{tz}{\sqrt{\alpha}}\right)}\right)\\
&=\lim_{\alpha\to\infty}\exp(-t\sqrt\alpha)\exp\left(\alpha(\operatorname{e}^{t/\sqrt\alpha}-1)\right)\\
&=\lim_{\alpha\to\infty}\exp\left(-t\sqrt\alpha +\alpha\left[t\alpha^{-1/2}+\frac{t^2\alpha^{-1}}{2}+\frac{t^3\alpha^{-3/2}}{6}+\cdots\right]\right)\\
&=\lim_{\alpha\to\infty}\exp\left(\frac{t^2}{2}+\frac{t^3\alpha^{-3/2}}{6}+\cdots\right)\\
&=\exp\left(\frac{t^2}{2}\right) \sim  N(0,1)
\end{align}$$


this result usually works very well if $\alpha > 20$, if we do not want to scale the variables then $Po(x|\alpha) \sim N(\alpha, \alpha)$

# method : 
The central limit theorem says that equally-weighted averages of samples from any distribution themselves are normally distributed. Consider the sample mean of independent and identically distributed random variables $X_1, X_2 . . . X_n \,$ such that $\,E[X_i] = μ \,$ and $ \,Var(X_i) = σ^2\,$:

from a statistical perspective the central limit theorem states that $\bar{X} \sim N(\mu, \frac{\sigma^2}{n})$

```python
#creating a set of axes with 100 bins, kernel density line set as true, color = red
N = 10000

data_poisson = poisson.rvs(loc = 4, size = N, mu=4)

ax = sns.histplot(data_poisson,
                  bins=30,
                  kde= True,
                  color='skyblue',
                  linewidth = 15,
                  alpha = 1)
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
```
