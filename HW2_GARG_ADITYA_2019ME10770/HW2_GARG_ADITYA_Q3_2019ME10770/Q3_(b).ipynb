```python
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def beta_plot(r, a1, b1, a2, b2, p1, p2, cdf=False, **kwargs):

    x = r # range of the x_values 
    
    # first we will check if we need the PDF or the CDF :) 
    
    if cdf: 
        y = (p1*beta.cdf(x, a1, b1)) +  (p2*beta.cdf(x, a2, b2)) # y values for the CDF
    
    else:
        y = (p1*beta.pdf(x, a1, b1)) +  (p2*beta.pdf(x, a2, b2)) # y values for the PDF
    
    plt.plot(x, y, **kwargs)
```

```python
import numpy as np 
```

```python
# dataset with H, T = 2, 3

plt.figure(figsize=(20,5))
x = np.linspace(0, 1, 50000)

y = beta_plot(x, a1 = 100, b1 = 100, a2 = 0.5, b2 = 0.5, p1 = 0.5, p2 = 0.5, color='red', label = 'prior', ls='--', lw=2)
y1 = beta_plot(x, a1 = 102, b1 = 103, a2 = 2.5, b2 = 3.5, p1 = 0.7258, p2 = 0.2742, color='blue', label = 'posterior')

plt.xlim(0,1)
plt.ylim(0,9)
plt.legend()
```
